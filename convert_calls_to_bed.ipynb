```python
#very simple script to convert the reformatted calls from "CNV_Analysis_plot" to bed files usable by ProOvErlap

import os
```

```python
files=['1000000_G1-2_reformat_calls.csv',
 '1000000_G1-3_reformat_calls.csv',
 '1000000_G1_reformat_calls.csv',
 '1000000_hepg2_reformat_calls.csv',
 '200000_G1-2_reformat_calls.csv',
 '200000_G1-3_reformat_calls.csv',
 '200000_G1_reformat_calls.csv',
 '200000_hepg2_reformat_calls.csv',]
```

```python
import pandas as pd
for file in files:
    file1 = pd.read_csv(file, index_col = 0)
    file1 = file1[['chr', 'start', 'end']]
    file1['name'] = '.'
    file1['score'] = "."
    file1['strand'] = '+'
    name = file.split('.')[0]+'.bed'
    file1.to_csv(name, index=False, header=False, sep='\t')

```
